# Задание 2

Рассчитать накопительный доход и накопительный ARPU для 37 месяцев после
активации пробного периода (от 0 до 36).

```python
cost_month_free = 6
cost_year_free = 35
```

```python
CTSP_month_free = 0.65
CTSP_year_free = 0.4
```

```python
year_free = [1, 0.27, 0.15, 0.07, 0.02]
```

```python
month_free = [1, 0.6, 0.45, 0.37, 0.3, 0.27, 0.23, 0.2, 0.17, 0.15, 0.13, 0.11, 0.09, 
              0.08, 0.06, 0.05, 0.04, 0.02, 0.01, 0.01]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def accumulate_for_month_free(conv, cost, CTSP, list):
    
    accumulate_income = 0
    list_for_viz_X =[]
    list_for_viz_Y = []
    
    for i in range(len(list)):
        accumulate_income += conv * cost * CTSP * list[i]
        print(f'Місяць №{i} — {round(accumulate_income, 2)}$')
        
        list_for_viz_X.append(i)
        list_for_viz_Y.append(accumulate_income)
    
    print(f'Сумарно за {len(list)} місяців — {round(accumulate_income, 2)}$')
    
    fig, ax = plt.subplots(figsize=(10, 6))
    scatter = ax.scatter(list_for_viz_X, list_for_viz_Y)
    
    ax.set(title='Accumulate income for month free', xlabel='Months', ylabel='Dollars')
    
    return accumulate_income
```

```python
Baseline_month_accumulate_income = accumulate_for_month_free(Baseline_month_free, cost_month_free, 
                                                            CTSP_month_free, month_free)
```

```python
Variant_A_month_accumulate_income = accumulate_for_month_free(Variant_A_month_free, cost_month_free, 
                                                            CTSP_month_free, month_free)
```

```python
def accumulate_for_year_free(conv, cost, CTSP, list):
    
    accumulate_income = 0
    list_for_viz_X =[]
    list_for_viz_Y = []
    
    for i in range(len(list)):
        accumulate_income += conv * cost * CTSP * list[i]
        print(f'Рік №{i} — {round(accumulate_income, 2)}$')
        
        list_for_viz_X.append(i)
        list_for_viz_Y.append(accumulate_income)
        
    print(f'Сумарно за {len(list)} років — {round(accumulate_income, 2)}$')
    
    fig, ax = plt.subplots(figsize=(10, 6))
    scatter = ax.scatter(list_for_viz_X, list_for_viz_Y)
    
    ax.set(title='Accumulate income for year free', xlabel='Years', ylabel='Dollars')
    
    return accumulate_income
```

```python
Baseline_year_accumulate_income = accumulate_for_year_free(Baseline_year_free, cost_year_free, 
                                                            CTSP_year_free, year_free[:4])
```

```python
Variant_A_year_accumulate_income = accumulate_for_year_free(Variant_A_year_free, cost_year_free, 
                                                            CTSP_year_free, year_free[:4])
```

```python
Baseline_accumulate_income = Baseline_month_accumulate_income + Baseline_year_accumulate_income
Baseline_accumulate_income
```

```python
Variant_A_accumulate_income = Variant_A_month_accumulate_income + Variant_A_year_accumulate_income
Variant_A_accumulate_income
```

```python
Baseline_ARPU = round(Baseline_accumulate_income / Baseline_users, 2)
Baseline_ARPU
```

```python
Variant_A_ARPU = round(Variant_A_accumulate_income / Variant_A_users, 2)
Variant_A_ARPU
```
